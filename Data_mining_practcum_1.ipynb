**This case study demonstrates data preprocessing and cleaning on a movie dataset using Python. It identifies missing values, noisy budget data, inconsistent country names, and duplicate records, then applies cleaning techniques and LabelEncoder to transform categorical variables for data mining analysis.**

```python
import pandas as pd
```

```python
import numpy as np
```

```python
dataset = pd.read_csv('DATA PREPROCESS.csv')
dataset
```

```python
dataset.info()
```

```python
missing_data = dataset[dataset.isnull().any(axis=1)] #axis=1 as we find for column for raw it's 0
missing_data
```

```python
dataclean1 = dataset.dropna()
dataclean1
```

```python
selected_rows = dataclean1[dataclean1['MovieID'].isin([12, 14, 16])]
selected_rows
```

```python
noise_data = dataclean1[dataclean1['BudgetUSD']<100000]
noise_data
```

```python
dataclean1.loc[dataclean1['MovieID'] == 21, 'BudgetUSD'] = 273022.95
dataclean1.loc[dataclean1['MovieID'] == 33, 'BudgetUSD']=4171378
dataclean1.loc[dataclean1['MovieID'] == 1360, 'BudgetUSD']=1203401.48
dataclean1.loc[dataclean1['MovieID'] == 1944, 'BudgetUSD']=330023.48
```

```python
selected_rows = dataclean1[dataclean1['MovieID'].isin([21, 33,1360, 1944])]
selected_rows
```

```python
inconsistent_data = dataclean1[dataclean1['Country'].str.contains(r"[^a-zA-Z\s]",na=False)]
inconsistent_data
```

```python
dataclean1['Country']=dataclean1['Country'].str.replace(r"[^a-zA-Z]","",regex=True)
dataclean1
```

```python
# Duplicat data
duplicate_data = dataclean1[dataclean1.duplicated(keep=False)]
duplicate_data
```

```python
dataclean2 = dataclean1.drop_duplicates()
dataclean2
```

```python
select_row = dataclean2[dataclean2['MovieID'].isin([24])]
select_row
```

```python
#transform the data
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataclean2['Country'] = le.fit_transform(dataclean2['Country'])
dataclean2
```

```python
dataclean2['Genre'] = le.fit_transform(dataclean2['Genre'])
for i, genre in enumerate(le.classes_):
    print(f"{i}: {genre}")
dataclean2
```
