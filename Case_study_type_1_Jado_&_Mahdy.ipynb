***This project focuses on data preprocessing and cleaning using Python. The dataset is analyzed to detect missing values, noisy data (illogical ages), inconsistent text entries, and duplicate records. Data cleaning techniques such as median/mode imputation, text normalization, duplicate removal, and categorical encoding using LabelEncoder are applied to prepare the dataset for machine learning or data analysis.***

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
# Load dataset
dataset = pd.read_csv('[IP] DATASET SC PRE-PROCESSING TYPE 1.csv')
dataset
```

```python
# Display dataset info
dataset.info()
```

```python
# --- STEP B: IDENTIFY MISSING DATA ---
missing_data = dataset[dataset.isnull().any(axis=1)]   # axis=1 → row-wise check
missing_data
```

```python
# Drop missing data (alternative cleaning version)
dataclean1 = dataset.dropna()
dataclean1
```

```python
# --- STEP C: CHECK NOISE (ILLOGICAL OR WRONG AGE VALUES) ---
# Example: Find users with illogical ages (below 10 or above 100)
noise_data = dataclean1[(dataclean1['age'] < 10) | (dataclean1['age'] > 100)]
noise_data
```

```python
# Replace illogical ages with median
median_age = dataclean1['age'].median()
dataclean1.loc[dataclean1['age'] < 10, 'age'] = median_age
dataclean1.loc[dataclean1['age'] > 100, 'age'] = median_age
dataclean1
```

```python
# --- STEP D: INCONSISTENT DATA (Subscription Type Text Variation) ---
# Check inconsistent entries
inconsistent_data = dataclean1[~dataclean1['subscription_type'].isin(['Free','Premium','Student','Family'])]
inconsistent_data
```

```python
# Fix text inconsistency (capitalize first letter)
dataclean1['subscription_type'] = dataclean1['subscription_type'].astype(str).str.strip().str.title()
dataclean1
```

```python
# --- STEP E: DUPLICATE DATA ---
duplicate_data = dataclean1[dataclean1.duplicated(keep=False)]
duplicate_data
```

```python
# Remove duplicate data
dataclean2 = dataclean1.drop_duplicates()
dataclean2
```

```python
# --- STEP F: TRANSFORM CATEGORICAL TO NUMERIC (ENCODING) ---
le = LabelEncoder()
```

```python
# Encode 'subscription_type'
dataclean2['subscription_type'] = le.fit_transform(dataclean2['subscription_type'])
dataclean2
```

```python
# Print label encoding mapping
for i, label in enumerate(le.classes_):
    print(f"{i}: {label}")
```

```python
# --- STEP G: FINAL CLEANED DATASET OUTPUT ---
dataclean2
```

```python
# Save cleaned data (optional)
dataclean2.to_csv('CLEANED_[IP]_DATASET.csv', index=False)
```

```python
# --- PART A: IDENTIFY DATA ISSUES ---

import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

# Load dataset
dataset = pd.read_csv('[IP] DATASET SC PRE-PROCESSING TYPE 1.csv')
dataset

# Dataset info
dataset.info()

# Missing values
missing_data = dataset[dataset.isnull().any(axis=1)]
missing_data

# Drop missing data (optional clean version)
dataclean1 = dataset.dropna()
dataclean1

# Detect noisy (illogical) ages
noise_data = dataclean1[(dataclean1['age'] < 10) | (dataclean1['age'] > 100)]
noise_data

# Detect inconsistent subscription types
inconsistent_data = dataclean1[~dataclean1['subscription_type'].isin(['Free','Premium','Student','Family'])]
inconsistent_data

# Detect duplicates
duplicate_data = dataclean1[dataclean1.duplicated(keep=False)]
duplicate_data
```

```python
# --- PART B: DATA CLEANING & TRANSFORMATION ---

# Replace missing values with median/mode
dataset['age'].fillna(dataset['age'].median(), inplace=True)
dataset['gender'].fillna(dataset['gender'].mode()[0], inplace=True)
dataset['country'].fillna(dataset['country'].mode()[0], inplace=True)
dataset['subscription_type'].fillna('Free', inplace=True)
dataset['listening_time'].fillna(dataset['listening_time'].median(), inplace=True)

# Remove duplicates
dataset = dataset.drop_duplicates()

# Fix inconsistent text format
dataset['subscription_type'] = dataset['subscription_type'].astype(str).str.strip().str.title()

# Fix illogical ages
median_age = dataset['age'].median()
dataset.loc[(dataset['age'] < 10) | (dataset['age'] > 100), 'age'] = median_age

# Transform categorical data into numeric
le = LabelEncoder()
dataset['subscription_type'] = le.fit_transform(dataset['subscription_type'])

# Show cleaned dataset
dataset.head()

# Save cleaned file (optional)
dataset.to_csv('CLEANED_[IP]_DATASET.csv', index=False)
print("✅ Cleaned dataset saved successfully!")
```
